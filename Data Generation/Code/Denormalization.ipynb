```python
def full_to_short_form(text):
    txt = ""
    for line in text:
        for expansion, contraction in contractions.items():
            line = line.replace(expansion, contraction)
        txt += line
    return txt
```

```python
def num_to_text(text):
    num_text = ""
    for i in range(len(text)):
        line = text[i]
        words = line.split()
        sent = ""
        for word in words:
            if word == "1st":
                word = "first"
            elif word == "2nd":
                word = "second"
            elif word == "3rd":
                word = "third"
            elif word == "4th":
                word = "fourth"
            elif word == "5th":
                word = "fifth"
            elif word == "6th":
                word = "sixth"
            elif word == "7th":
                word = "seventh"
            elif word == "8th":
                word = "eighth"
            elif word == "9th":
                word = "ninth"
            elif word.isnumeric():
                try:
                    word = num2words(word)
                except:
                    word = word
            sent += word + " "
        num_text += sent + "\n"
    return num_text
```

```python
with open("Only_English_train_Norm.txt") as f:
    cs_text = f.readlines()
f.close()

num_text = full_to_short_form(num_to_text(cs_text))

f2 = open("Only_English_train_Normalized.txt",'w+')
f2.write(num_text)
f2.close()

print("Done")
```
