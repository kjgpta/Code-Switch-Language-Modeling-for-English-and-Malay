```python
def adding_lang_tag(filepath,filetosavetext):
    with open(filepath) as f:
        data = f.readlines()
    f.close()
    phrase_data = ""
    for line in data:
        list_phrase = []
        malay_flag = False
        eng_flag = False
        other_flag = False
        inter_flag = False
        phrase = ""
        words = line.split()
        if words[0][0] == "#":
            other_flag = True
        elif words[0][0] == "!":
            inter_flag = True
        elif words[0].startswith("<malay>"):
            malay_flag = True
        else:
            eng_flag = True
        for i in words:
            if other_flag or i[0] == "#" or i[0] == "!":
                if phrase != "" and not other_flag:
                    list_phrase.append("<english>" + phrase[:-1] + "</english>")
                    phrase = ""
                    eng_flag = False
                other_flag = True
                if (i[0] == "#" and i[-1] == "#"):
                    phrase = ""
                    phrase += i[1:-1]
                    list_phrase.append("<other_language>" + phrase + "</other_language>")
                    phrase = ""
                    other_flag = False
                elif i[0] == "#":
                    phrase = ""
                    phrase += i[1:] + " "
                elif i[-1] == "#":
                    phrase += i[:-1]
                    list_phrase.append("<other_language>" + phrase + "</other_language>")
                    phrase = ""
                    other_flag = False
                elif other_flag:
                    phrase += i + " "
            elif inter_flag or i[0] == "!":
                if phrase != "" and not other_flag:
                    list_phrase.append("<english>" + phrase[:-1] + "</english>")
                    phrase = ""
                    eng_flag = False
                inter_flag = True
                if (i[0] == "!" and i[-1] == "!"):
                    phrase = ""
                    phrase += i[1:-1]
                    list_phrase.append("<interjection>" + phrase + "</interjection>")
                    phrase = ""
                    inter_flag = False
                elif i[0] == "!":
                    phrase = ""
                    phrase += i[1:] + " "
                elif i[-1] == "!":
                    phrase += i[:-1]
                    list_phrase.append("<interjection>" + phrase + "</interjection>")
                    phrase = ""
                    inter_flag = False
                elif inter_flag:
                    phrase += i + " "
            elif malay_flag or i.startswith("<malay>"):
                if phrase != "" and not malay_flag:
                      list_phrase.append("<english>" + phrase[:-1] + "</english>")
                      phrase = ""
                      eng_flag = False
                malay_flag = True
                if i.startswith("<malay>") and i.endswith("</malay>"):
                    phrase = ""
                    phrase += i[7:-8]
                    list_phrase.append("<malay>" + phrase + "</malay>")
                    phrase = ""
                    malay_flag = False
                elif i.startswith("<malay>"):
                    phrase = ""
                    phrase += i[7:] + " "
                elif i.endswith("</malay>"):
                    phrase += i[:-8]
                    list_phrase.append("<malay>" + phrase + "</malay>")
                    phrase = ""
                    malay_flag = False
                elif malay_flag:
                    phrase += i + " "
            else:
                eng_flag = True
                phrase += i + " "
        if phrase != "":
            list_phrase.append("<english>" + phrase[:-1] + "</english>")
            phrase = ""
            eng_flag = False
        for j in list_phrase:
            phrase_data += j + " "
        phrase_data += "\n"
    f1 = open(filetosavetext,'w+')
    f1.write(phrase_data)
    f1.close()  
```
